```julia
rmsd_mean = 0.0
icount = 0
for i = 1:size(t, 1)
    for j = (i+1):size(t, 1)
        t1 = t[i, :]
        t2 = superimpose(t1, t[j, :])
        rmsd_mean += compute_rmsd(t1, t2)[1]
        icount += 1
    end
end
rmsd_mean / icount
```
